```python
import numpy as np
import matplotlib.pyplot as plt

img = plt.imread('flower.tiff')
print(img.shape)
print(np.min(img), np.max(img))
```

```python
plt.imshow(img)
```

```python
# RGB
img_red = img[:,:, 0]
img_green = img[:,:, 1]
img_blue = img[:,:, 2]

plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.imshow(img_red, 'gray')
plt.subplot(1, 3, 2)
plt.imshow(img_green, 'gray')
plt.subplot(1, 3, 3)
plt.imshow(img_blue, 'gray')
```

```python
plt.imshow(img[:,:,3], 'gray')
print(np.min(img[:,:,3]), np.max(img[:,:,3]))
```

```python
img_red2 = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
img_red2[:,:,0] = img[:,:, 0]

img_green2 = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
img_green2[:,:,1] = img[:,:, 1]

img_blue2 = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
img_blue2[:,:,2] = img[:,:, 2]

plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.imshow(img_red2)
plt.subplot(1, 3, 2)
plt.imshow(img_green2)
plt.subplot(1, 3, 3)
plt.imshow(img_blue2)
```

```python
def rgb2gray(img):
    
    img_gray = 0.299*img[:,:,0] + 0.587*img[:,:,1] + 0.114*img[:,:,2]
    return img_gray

img_gray = rgb2gray(img)
plt.imshow(img_gray, 'gray')
```

# Formatos de imagem

```python
def image_info(img):
    
    print(f'Tamanho: {img.shape}\n',
          f'Tipo de dado: {img.dtype}\n',
          f'Menor valor: {img.min()}\n',
          f'Maior valor: {img.max()}\n',
          f'Valores únicos: {len(np.unique(img))}',
          sep=''
         )
```

```python
img = plt.imread('flower.tiff')
print('tiff')
image_info(img)

img = plt.imread('flower.jpg')
print('\njpg')
image_info(img)

img = plt.imread('flower.png')
print('\npng')
image_info(img)
```

```python
img_255 = np.round(255*img).astype(np.uint8)
image_info(img_255)
```

```python
img1 = plt.imread('flower_grayscale1.tiff')
print('img1')
image_info(img1)

img2 = plt.imread('flower_grayscale2.tiff')
print('\nimg2')
image_info(img2)
```

# Mapeamento do matplotlib

```python
img = np.zeros([100, 300])

img[25:75, 25:75] = 20
img[25:75, 125:175] = 40
img[25:75, 225:275] = 60
plt.imshow(img, 'gray')
plt.colorbar(orientation='horizontal')
```

```python
plt.imshow(img, 'gray', vmin=0, vmax=255)
plt.colorbar(orientation='horizontal')
```

```python
img = np.zeros([100, 300, 3], dtype=np.uint8)

img[25:75, 25:75, 0] = 60
img[25:75, 125:175, 1] = 60
img[25:75, 225:275, 2] = 60
plt.imshow(img)
```

```python
img = plt.imread('flower_grayscale2.tiff')
img = img.astype(float)
img_log = np.log(img+1)
image_info(img_log)
```

```python
def renormalize(img):
    
    img_norm = img - np.min(img)
    img_norm = img_norm/np.max(img_norm)
    img_norm = 255*img_norm
    img_norm = img_norm.astype(np.uint8)
    
    return img_norm

img_norm = renormalize(img)
image_info(img_norm)
```
