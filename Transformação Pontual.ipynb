```python
img = plt.imread('cameraman.tiff')
plt.imshow(img, cmap='gray')
```

# Transformações de imagens

### Negativo da imagem

```python
num_linhas, num_cols = img.shape

#img negativa em python puro
img_neg = np.zeros([num_linhas, num_cols], dtype = np.uint8)
for row in range(num_linhas):
    for col in range(num_cols):
        img_neg[row, col] = 255 - img[row, col]
    
plt.imshow(img_neg, 'gray')
```

```python
#img neg usando o numpy, que é bem mais eficiente por ser 
#implementado em C

img_neg_np = 255 - img
plt.imshow(img_neg_np, 'gray')
```

### Logaritmo da imagem

```python
img_log = np.zeros((num_linhas, num_cols), dtype=np.float)
for row in range(num_linhas):
    for col in range(num_cols):
        img_log[row, col] = np.log(1+img[row, col])
plt.imshow(img_log, 'gray')
```

```python
x = range(1, 255)
y = np.log(x)
plt.plot(x, y)
```

```python
#usando lookuptable
lookup_table = np.zeros(256)
for value in range(0, 256):
    lookup_table[value] = np.log(1+value)

img_log = np.zeros((num_linhas, num_cols), dtype=np.float)
for row in range(num_linhas):
    for col in range(num_cols):
        img_log[row, col] = lookup_table[img[row, col]]
plt.imshow(img_log, 'gray')
```
