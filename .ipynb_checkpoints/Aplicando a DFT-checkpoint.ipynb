```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq, fftshift
```

```python
freq_s1 = 2
freq_s2 = 5
x = np.arange(0, 3, 0.01)
y = 8*np.cos(2*np.pi*freq_s1*x) + 7*np.cos(2*np.pi*freq_s2*x) + 3*np.random.normal(0.,1.,x.size)
plt.plot(x, y)
```

```python
Fy = fft(y)
freqs = fftfreq(len(x), x[1]-x[0])
Fy = fftshift(Fy)
freqs = fftshift(freqs)
plt.plot(freqs, np.abs(Fy))
```

```python
x = np.linspace(0, 1, 100)
y = np.zeros(len(x))
y[45:55] = 1

Fy = fft(y)
freqs = fftfreq(len(x), x[1]-x[0])
Fy = fftshift(Fy)
freqs = fftshift(freqs)

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.subplot(1, 2, 2)
plt.plot(freqs, np.abs(Fy))
```

```python
from scipy.fftpack import fft2

N = 150
freq_r = 2./N
freq_c = 5./N
z = np.zeros((N, N))

for row in range(N):
    for col in range(N):
        z[row, col] = 4*np.cos(2*np.pi*freq_r*row)+5*np.cos(2*np.pi*freq_c*col)

plt.imshow(z, 'gray')
```

```python
Fz = fft2(z)
freqs = fftfreq(N, 1.)
Fz = fftshift(Fz)
freqs = fftshift(freqs)

plt.pcolormesh(freqs, freqs, np.abs(Fz), cmap='gray', shading='auto')
plt.xlim((-0.1, 0.1))
plt.ylim((-0.1, 0.1))
```

```python
def calc_fourier_and_plt(img):
    
    num_rows, num_cols = img.shape
    
    Fing = fft2(img)
    freq_r = fftfreq(num_rows)
    freq_c = fftfreq(num_cols)
    
    Fing = fftshift(Fing)
    freq_r = fftshift(freq_r)
    freq_c = fftshift(freq_c)
    plt.figure(figsize=[8,8])
    plt.pcolormesh(freq_c, freq_r, np.log(np.abs(Fing)), cmap='gray', shading='auto')
```

```python
img = plt.imread('imgs/tiles1.tiff')
plt.figure(figsize=[8, 8])
plt.imshow(img, cmap='gray')
calc_fourier_and_plt(img)
```
