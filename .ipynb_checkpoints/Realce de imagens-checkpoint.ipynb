```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve
import scipy.ndimage as ndi

img = plt.imread('imgs/moon.tiff')
img_float = img.astype(float)
plt.figure(figsize=[7,7])
plt.imshow(img, 'gray')
```

```python
filtro_lap = np.array([[0, -1, 0],
                       [-1, 4, -1],
                       [0, -1, 0]])
img_smooth = ndi.gaussian_filter(img_float, sigma=1)
img_lap = convolve(img_smooth, filtro_lap, 'same')
plt.figure(figsize=[7,7])
plt.imshow(img_lap, 'twilight')
```

```python
c = 5
img_improv = img_float + c*img_lap
plt.figure(figsize=[12, 6])
plt.subplot(1, 2, 1)
plt.imshow(img, 'gray')
plt.subplot(1, 2, 2)
plt.imshow(img_improv, 'gray')
```

```python
img_smooth = ndi.gaussian_filter(img_float, sigma=5)
img_diff = img_float - img_smooth
plt.figure(figsize=[7,7])
plt.imshow(img_diff, 'twilight')
```

```python
c = 1
img_boosted = img_float + c*img_diff
plt.figure(figsize=[12, 6])
plt.subplot(1, 2, 1)
plt.imshow(img, 'gray')
plt.subplot(1, 2, 2)
plt.imshow(img_boosted, 'gray')
```
