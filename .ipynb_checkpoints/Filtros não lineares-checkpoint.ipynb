# Média geométrica para array

\begin{equation}
Geometric mean = \sqrt[n]{x_1 * x_2 * x_3 * ... * x_n}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

img_1D = np.array([[10, 40, 70, 230, 200, 100, 10, 20, 30, 160, 200, 40]])
# img_1D = np.array([[10, 40, 70, 230, 200, 100, 10]])
plt.imshow(img_1D, 'gray', vmin=0, vmax=255)
```

```python
# função de média geométrica 1D com filtro de tamanho n
def geometric_mean_1D(signal, n):

    # ignora a linha da matriz e pega o numero de colunas
    _, num_cols = signal.shape

    # define um sinal com padding em volta e um sinal de saida
    # signal_padded é preenchido com valores 1, neutros na multiplicação
    signal_padded = np.full((1, num_cols + n-1), 1, dtype="float64")
    signal_filtered = np.zeros((1, num_cols))

    # salva o resultado de n//2 por praticidade
    offset = n//2

    # preenche o signal_padded com a imagem original
    for col in range(num_cols):
        signal_padded[0, col + offset] = signal[0][col]

    # percorre as posições 
    for col in range(offset, num_cols + offset):

        # inicia o produtório com o pixel atual
        produtorio = signal_padded[0, col]

        # para cada posição do filtro, adiciona ao produtório os valores vizinhos do pixel
        for filter_col in range (1, offset + 1):
            produtorio *= signal_padded[0, col - filter_col] * signal_padded[0, col + filter_col]

        # salva o resultado na posição correspondente do sinal filtrado
        signal_filtered[0, col - offset] = produtorio**(1/n)
    
    return signal_filtered
```

```python
img_1D_filtered = geometric_mean_1D(img_1D, 3)

plt.figure(figsize=[13, 4])

plt.subplot(2, 1, 1)
plt.title("Original")
plt.imshow(img_1D, 'gray', vmin=0, vmax=255)

plt.subplot(2, 1, 2)
plt.title("Filtrada")
plt.imshow(img_1D_filtered, 'gray', vmin=0, vmax=255)

print(img_1D)
print(img_1D_filtered)
```

# Média geométrica para imagem 2D
\begin{equation}
G(x, y) = \left[ \prod_{i, j \in S} S(i, j) \right]^{\frac{1}{mn}}
\end{equation}

```python
# função de média geométrica 2D com filtro de tamanho S
def geometric_mean(img, S):

    num_rows, num_cols = img.shape
    
    half_r_size = S//2
    half_c_size = S//2

    # img_padded corresponde a imagem original com padding em volta
    # o preenchimento é feito com os valores dos cantos da imagem original
    img_padded = np.pad(img, ((half_r_size,half_r_size), (half_c_size,half_c_size)), mode='edge')
    plt.imshow(img_padded,'gray')

    # define uma imagem de saída
    img_filtered = np.zeros((num_rows, num_cols))

    for row in range(num_rows):
        for col in range(num_cols):
            # termo neutro da multiplicação
            prod_region = 1
            
            # percorre uma região equivalente ao tamanho do filtro
            for f_row in range(S):
                for f_col in range(S):
                    prod_region *= int(img_padded[row + f_row, col + f_col])
            
            # define o produto obtido para uma região específica
            prod_region = prod_region ** (1 / (S**2))
            
            # salva o resultado na posição correspondente do sinal filtrado
            img_filtered[row, col] = prod_region
            
    return img_filtered
```

```python
img = plt.imread('imgs/cameraman.tiff')

# imagem original
plt.imshow(img, cmap='gray')
```

```python
plt.figure(figsize=[20, 10])

# imagem com filtro de tamanho 3
plt.subplot(1, 3, 1)
img_filtered = geometric_mean(img, 3)
plt.imshow(img_filtered, 'gray')

# imagem com filtro de tamanho 5
plt.subplot(1, 3, 2)
img_filtered = geometric_mean(img, 5)
plt.imshow(img_filtered, 'gray')

# imagem com filtro de tamanho 11
plt.subplot(1, 3, 3)
img_filtered = geometric_mean(img, 11)
plt.imshow(img_filtered, 'gray')
```

# Mediana para array

\begin{equation}
f(x, y) = mediana_{(s, t) \in S_{x, y}}\left[ f(s, t) \right]
\end{equation}

```python
img_1D = np.array([[10, 40, 70, 230, 200, 100, 10, 20, 30]])
plt.imshow(img_1D, 'gray', vmin=0, vmax=255)
```

```python
# função de mediana 1D com filtro de tamanho n
def median_1D(signal, n):

    # ignora a linha da matriz e armazena apenas o numero de colunas
    _, num_cols = signal.shape
    
    # salva o resultado de n//2 por praticidade
    offset = n//2
    
    # define um sinal com padding em volta e um sinal de saida
    signal_padded = np.pad(signal, ((0,), (offset,)), mode='symmetric')
    signal_filtered = np.zeros((1, num_cols))
    
    
    # define um array auxiliar para calculo da mediana
    aux_median = np.zeros((1, n))
    
    # percorre todas as colunas
    for col in range(num_cols):

        # armazena as colunas selecionadas pelo filtro
        aux_median = signal_padded[0, col:col+n]

        # realiza ordenacao crescente
        aux_median = np.sort(aux_median)
        
        # armazena o valor contido no centro do array ordenado
        median_value = aux_median[offset]
        
        # preenche o sinal filtrado
        signal_filtered[0, col] = median_value
        
    return signal_filtered
```

```python
img_1D_filtered = median_1D(img_1D, 3)

plt.figure(figsize=[13, 4])

plt.subplot(2, 1, 1)
plt.title("Original")
plt.imshow(img_1D, 'gray', vmin=0, vmax=255)

plt.subplot(2, 1, 2)
plt.title("Filtrada")
plt.imshow(img_1D_filtered, 'gray', vmin=0, vmax=255)

print(img_1D)
print(img_1D_filtered)
```

# Mediana para imagem 2D

```python
# função de mediana 2D com filtro de tamanho n
def median_2D(img, n):
    
    num_rows, num_cols = img.shape
    f_size = n*n

    # salva o resultado de f_size//2 por praticidade
    mediana = f_size//2

    half_r_size = n//2
    half_c_size = n//2

    # img_padded corresponde a imagem original com padding em volta
    # o preenchimento da borda é feito simetricamente
    img_padded = np.pad(img, ((half_r_size, half_r_size), (half_c_size, half_c_size)), mode='symmetric')

    # array que conterá os valores da regiao explorada
    median_array = np.zeros((1, f_size))

    # imagem de saida
    img_filtered = np.zeros((num_rows, num_cols))

    for row in range(num_rows):
        for col in range(num_cols):

            # percorre uma regiao equivalente ao tamanho do filtro
            for f_row in range(n):
                for f_col in range(n):

                    median_array[0][f_col + f_row*(n-1)] = int(img_padded[row + f_row, col + f_col])

            # realiza ordenacao crescente e salva o valor contido no centro do array
            median_array = np.sort(median_array)
            median_value = median_array[0,mediana]

            # salva o resultado na posição correspondente do sinal filtrado
            img_filtered[row, col] = median_value

    return img_filtered
```

```python
img_filtered = median_2D(img, 10)

plt.figure(figsize=[11,11])

plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img, 'gray')

plt.subplot(1, 2, 2)
plt.title("Filtrada")
plt.imshow(img_filtered, 'gray')
```
