```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 40, 100)
fx = x**2
plt.plot(x, fx)
```

```python
dx = x[1] - x[0]
dfx = np.zeros(len(fx)-1)

for i in range(len(fx)-1):
    dfx[i] = (fx[i+1]-fx[i])/dx
    
plt.plot(x[:-1], dfx)
```

```python
#mesma coisa fazendo com o numpy
dfx = (fx[1:] - fx[:-1])/dx
#outra opção
dfx = np.diff(fx)/dx
plt.plot(x[:-1], dfx)
```

```python
x = np.linspace(0, 4*np.pi, 101)
dx = x[1] - x[0]
fx = np.sin(x)
dfx = np.diff(fx)/dx
plt.plot(x, fx)
plt.figure()
plt.plot(x[:-1], dfx)
```

```python
x = np.linspace(0, 1, 100)
fx_complex = 5 + x**2 + np.sqrt(x) - 2*x**3
dx = x[1] - x[0]
dfx_complex = np.diff(fx_complex)/dx

plt.plot(x, fx_complex)
plt.figure()
plt.plot(x[:-1], dfx_complex)
```

```python
img = plt.imread("imgs/cameraman.tiff")
img = img.astype(int)

num_rows, num_cols = img.shape
img_dx = np.zeros((num_rows, num_cols-1))
for row in range(num_rows):
    for col in range(num_cols-1):
        img_dx[row, col] = img[row, col+1] - img[row, col]

img_dy = np.zeros((num_rows-1, num_cols))
for col in range(num_cols):
    for row in range(num_rows-1):
        img_dy[row, col] = img[row+1, col] - img[row, col]

plt.figure(figsize=[15, 5])
plt.subplot(1, 3, 1)
plt.imshow(img, 'twilight_shifted')
plt.subplot(1, 3, 2)
plt.imshow(img_dx, 'twilight_shifted')
plt.subplot(1, 3, 3)
plt.imshow(img_dy, 'twilight_shifted')
```

```python
plt.figure(figsize=[15, 5])
plt.subplot(1, 3, 1)
_ = plt.hist(img.flatten(), 50)
plt.subplot(1, 3, 2)
_ = plt.hist(img_dx.flatten(), 50)
plt.subplot(1, 3, 3)
_ = plt.hist(img_dy.flatten(), 50)
```

```python
img_grad_mag = np.zeros((num_rows-1, num_cols-1))
for row in range(num_rows-1):
    for col in range(num_cols-1):
        img_grad_mag[row, col] = np.sqrt(img_dx[row, col]**2 + img_dy[row, col]**2)

plt.figure(figsize=[7,7])
plt.imshow(img_grad_mag, 'twilight_shifted')
```

```python
from scipy.signal import convolve

##fazer a concolução com esse filtro é 
#o equivalente a correlaçao com filtro [[-1, 1]] q vimos em aula
wx = np.array([[1, -1]]) 
wy = np.array([[1], 
               [-1]])

img_dx_conv = convolve(img, wx, mode='same')
img_dy_conv = convolve(img, wy, mode='same')

plt.figure(figsize=[15, 5])
plt.subplot(1, 3, 1)
plt.imshow(img, 'gray')
plt.subplot(1, 3, 2)
plt.imshow(img_dx_conv, 'gray')
plt.subplot(1, 3, 3)
plt.imshow(img_dy_conv, 'gray')
```

```python
sx = np.array([[1, 0, -1],
               [2, 0, -2],
               [1, 0, -1]])
sy = sx.T #transposta de sx

img_dx_conv_sobel = convolve(img, sx, mode='same')
img_dy_conv_sobel = convolve(img, sy, mode='same')

plt.figure(figsize=[15, 5])
plt.subplot(1, 3, 1)
plt.imshow(img, 'gray')
plt.subplot(1, 3, 2)
plt.imshow(img_dx_conv_sobel, 'gray')
plt.subplot(1, 3, 3)
plt.imshow(img_dy_conv_sobel, 'gray')
```

## Visualização do gradiente

```python
def view_gradient(img, img_dx, img_dy, region):
    
    Y, X = np.mgrid[region[0]:region[1], region[2]:region[3]]
    Y = Y[::-1]
    img_region = img[region[0]:region[1], region[2]:region[3]]
    img_dx_region = img_dx[region[0]:region[1], region[2]:region[3]]
    img_dy_region = img_dy[region[0]:region[1], region[2]:region[3]]
    
    plt.figure(figsize=[12, 6])
    plt.subplot(1, 2, 1, aspect='equal')
    plt.pcolormesh(X, Y, img_region, shading='auto', cmap='gray')
    plt.subplot(1, 2, 2, aspect='equal')
    plt.quiver(X, Y, img_dx_region, -img_dy_region, edgecolor='k', facecolor='k', 
               linewidth=.5, scale=3e3)

region = [55,70,100,120]    # [first row, last row, first column, last column]
view_gradient(img, img_dx_conv, img_dy_conv, region)
```

## Derivada com ruído

```python
def filtro_gaussiano_1d(S):
    
    sigma = S/6
    x = np.linspace(-3*sigma, 3*sigma, S)
    y = np.exp(-x**2/(2*sigma**2))
    y = y/np.sum(y)
    
    return y
```

```python
qnt_ruido = 0.3
x = np.linspace(0, 4*np.pi, 101)
fx = np.sin(x) + qnt_ruido*(np.random.rand(len(x))-0.5)
plt.plot(x, fx)
```

```python
fil_der = np.array([1, -1])
dfx = convolve(fx, fil_der, mode='same')
plt.plot(x, dfx)
```

```python
fil_suav = filtro_gaussiano_1d(25)
fx_suav = convolve(fx, fil_suav, mode='same')
dfx = convolve(fx_suav, fil_der, mode='same')
plt.plot(x, dfx)
```
